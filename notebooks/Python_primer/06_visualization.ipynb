### Two-Dimensional Data Set

```python
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y, lw=1.5)
plt.plot(y, 'go')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple Plot Example');
```

```python
plt.figure(figsize=(10, 4))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y, 'go')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple Plot Example');
```

```python
y[:, 0] = y[:, 0] * 100
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple Plot Example');
```

```python
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=8)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('Simple Plot Example')
ax2 = ax1.twinx()
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd');
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(211)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'rd')
plt.legend(loc=0)
plt.ylabel('value')
plt.title('Simple Plot Example')
plt.subplot(212)
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value');
```

```python
def two_data_sets_plot(): 
    plt.figure(figsize=(10, 6))
    plt.subplot(121)
    plt.plot(y[:, 0], lw=1.5, label='1st')
    plt.plot(y[:, 0], 'rD')
    plt.legend(loc=0)
    plt.xlabel('index')
    plt.ylabel('value')
    plt.title('1st Data Set')
    plt.subplot(122)
    plt.bar(np.arange(len(y)), y[:, 1], width=0.5,
            color='g', label='2nd')
    plt.legend(loc=0)
    plt.xlabel('index')
    plt.title('2nd Data Set');
```

```python
two_data_sets_plot()
```

## Static 3D Plotting

```python
strike = np.linspace(50, 150, 24)
```

```python
ttm = np.linspace(0.5, 2.5, 24)
```

```python
strike, ttm = np.meshgrid(strike, ttm)
```

```python
strike[:2].round(1)
```

```python
iv = (strike - 100) ** 2 / (100 * strike) / ttm
```

```python
iv[:5, :3]
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig, ax = plt.subplots(subplot_kw={'projection': '3d'}, figsize=(8, 8))
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2,
                       cmap=plt.cm.coolwarm, linewidth=0.5,
                       antialiased=True)
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
fig.colorbar(surf, shrink=0.5, aspect=5);
```

```python
def three_d_static_plot():
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.view_init(30, 60)
    ax.scatter(strike, ttm, iv, zdir='z', s=25,
               c='b', marker='^')
    ax.set_title('3-D Plot Example')
    ax.set_xlabel('strike')
    ax.set_ylabel('time-to-maturity')
    ax.set_zlabel('implied volatility');
```

```python
three_d_static_plot()
```

## Interactive 2D Plotting

### Basic Plots

```python
cf.set_config_file(offline=True)
```

```python
a = np.random.standard_normal((250, 5)).cumsum(axis=0)
```

```python
index = pd.date_range('2019-1-1',
                      freq='B',
                      periods=len(a)) # <4> 
```

```python
df = pd.DataFrame(100 + 5 * a,
                  columns=list('abcde'),
                  index=index)
```

```python
df.head()
```

```python
df.plot()
```

```python
df.iplot()
```

```python
df[['a', 'b']].iplot(
             theme='polar',
             title='A Time Series Plot',
             xTitle='date',
             yTitle='value',
             mode={'a': 'markers', 'b': 'lines+markers'},
             symbol={'a': 'cross', 'b': 'diamond'},
             size=3.5,
             colors={'a': 'blue', 'b': 'magenta'},
                        )
```

```python
df.iplot(kind='hist',
             subplots=True,
             bins=15,
        )
```

### Financial Plotting

```python
# data from FXCM Forex Capital Markets Ltd.
raw = pd.read_csv(PATH + 'fxcm_eur_usd_eod_data.csv',
                 index_col=0, parse_dates=True)
```

```python
raw.info()
```

```python
quotes = raw[['AskOpen', 'AskHigh', 'AskLow', 'AskClose']]
quotes = quotes.iloc[-60:]
quotes.tail()
```

```python
qf = cf.QuantFig(
         quotes,
         title='EUR/USD Exchange Rate',
         legend='top',
         name='EUR/USD'
)
```

```python
qf.iplot()
```

```python
qf.add_bollinger_bands(periods=15,
                       boll_std=2)
```

```python
qf.iplot()
```

```python
qf.add_rsi(periods=14,
          showbands=False)
```
